```python
import pandas as pd
from os import listdir
import numpy as np
```

```python
directory = "C:/Users/ricab/Desktop/data-x/Project/ead-model/data/ados_datasets/"
```

```python
filenames=listdir(directory)
```

```python
features=np.empty([0,0]);
for filename in filenames:
    txtname = directory+filename
    data = pd.read_csv(txtname, sep="\t", header=None)
    new_header=data.iloc[1]
    data = data[2:]
    data.columns = new_header
    f_features=new_header.unique()
    features=np.append(features, f_features)
    features=np.unique(features)
df=pd.DataFrame(columns=filenames, index=features)

for filename in filenames:
    txtname = directory+filename
    data = pd.read_csv(txtname, sep="\t", header=None)
    new_header=data.iloc[1]
    data = data[2:]
    f_features=new_header.unique()
    df.loc[f_features, filename]=1


```

```python
writer = pd.ExcelWriter('ColumnComparison.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='hoja1')
writer.save()
```
